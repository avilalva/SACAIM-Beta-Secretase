```python
import pandas as pd
from chembl_webresource_client.new_client import new_client

# Search for beta-secretase as the target
target = new_client.target
target_query = target.search('beta secretase')
targets_df = pd.DataFrame.from_dict(target_query)
targets_df[targets_df['organism'] == 'Homo sapiens']
```

```python
selected_target = targets_df.target_chembl_id[3]
selected_target
```

```python
activity = new_client.activity
result = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
df = pd.DataFrame.from_dict(result)
df
```

Handling Missing Data

```python
#function to check NaN values
def nan_check(dataframe,col_name):
  if dataframe[col_name].notna().any():
      print("NA is present")
  else:
      print("No NA values")

nan_check(df,'col_name')
```

```python
df.dropna(subset=['standard_value'], inplace=True)
nan_check(df)
df
# df_ic50_notna = df_ic50[df_ic50.standard_value.notna()]
# df_ic50_notna.standard_value
```

```python
selected_columns = df[['assay_chembl_id', 'standard_value']]
selected_columns
```

##**Preprocessing of Bio Activity data**

```python
desc = df['standard_value'].describe()
desc
```

```python
bioactivity_class = []
for i in df.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

```python
df_bioactivity_class = pd.DataFrame({'bioactivity_class': bioactivity_class})
# dataf.to_csv('bioactivity_3class.csv', index = False)
```

```python
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df3 = df[selection]
# df3.to_csv('df3.csv')
```

```python
df3 = df3.reset_index(drop=True)
df_bioactivity_class = df_bioactivity_class.reset_index(drop=True)
result_df = pd.concat([df3,df_bioactivity_class], axis=1)
```

```python
nan_check(result_df,'standard_value')
```

```python
result_df.to_csv('beta_secretase_bioact_data_preprocessed.csv')
```
